```python
from ultralytics import YOLO
import cv2
import pygame
import os

# Initialize pygame mixer
pygame.mixer.init()

model = YOLO('best.pt')

# Choose whether to use a video file or camera feed
use_video_file = True
if use_video_file:
    video_path = "rear_collision_10.mp4"  #stock-footage-a-uk-ambulance-on-a-road-drive-by-camera-filmed-york-yorkshire-uk.webm
    cap = cv2.VideoCapture(video_path)
else:
    # Use camera feed
    cap = cv2.VideoCapture(0)  # Use 0 for default camera, change if you have multiple cameras

# Load the alarm sound
alarm_sound = pygame.mixer.Sound("alarm.wav")

# Create a folder to save the video
output_folder = "C:\\Users\\UDITH\\Desktop\\final wala\\output_video"
os.makedirs(output_folder, exist_ok=True)

# Define the codec and create a VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'XVID')
output_video_path = os.path.join(output_folder, "output_video.avi")
output_video = cv2.VideoWriter(output_video_path, fourcc, 20.0, (int(cap.get(3)), int(cap.get(4))))

while cap.isOpened():
    # Read a frame from the video or camera
    success, frame = cap.read()

    if success:
        # Run YOLOv8 inference on the frame
        results = model(frame)

        for result in results:
            boxes = result.boxes.cpu().numpy()
            for box in boxes:
                if box.conf[0] > 0.6:
                    # Draw bounding box on the frame
                    frame = results[0].plot()

                    # Play the alarm sound
                    alarm_sound.play()

        # Write the frame with bounding boxes to the output video
        output_video.write(frame)

        # Display the frame with bounding boxes
        cv2.imshow("Video", frame)

        # Break the loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        # Break the loop if the end of the video is reached or camera feed is lost
        break

# Release the video capture object, close the display window, and release the VideoWriter object
cap.release()
output_video.release()
cv2.destroyAllWindows()
```
